```python

def get_track(s):
    i,moves,des,flag = s.split(' ')
    moves=moves.split(';')[:-1]
    moves=[move.split(',') for move in moves]
    moves=[[int(ll) for ll in l] for l in moves]
    des=[float(p) for p in des.split(',')]
    flag=int(flag)
    print(i,moves[0],'->',des,flag)
    m = np.array(moves)
    m = m[np.lexsort(m.T)]
#     print(m)
    return (i,(m[0,0],m[0,1]), m, des, flag)

def plot(data, rows=3, cols=3):
    for i,s in enumerate(data):
        if i >= rows*cols:
            break
        start, track, des, flag = get_track(s)
        canvas = plts[i//cols][i%cols]
        canvas.plot(*start, '.b')
        canvas.plot(track[:,0],track[:,1],'-g')
        canvas.plot(*des,'.r')
```

```python
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F

import torch.utils.data
import torchvision
import torchvision.datasets as dset
import torchvision.transforms as transforms
import torchvision.utils as vutils

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.ConvLayers = nn.Sequential(
            # input is (nc) x 64 x 64
            nn.Conv2d(nc, ndf, 4, 2, 1, bias=False),
            nn.LeakyReLU(0.2, inplace=True),
            # state size. (ndf) x 32 x 32
            nn.Conv2d(ndf, ndf * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ndf * 2),
            nn.LeakyReLU(0.2, inplace=True),
            # state size. (ndf*2) x 16 x 16
            nn.Conv2d(ndf * 2, ndf * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ndf * 4),
            nn.LeakyReLU(0.2, inplace=True),
            # state size. (ndf*4) x 8 x 8
            nn.Conv2d(ndf * 4, ndf * 8, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ndf * 8),
            nn.LeakyReLU(0.2, inplace=True),
            # state size. (ndf*8) x 4 x 4
            nn.Conv2d(ndf * 8, 1, 4, 2, 1, bias=False),
        )
        self.conv = nn.Conv2d(3, 1, 4, 2, 1, bias=False)
        self.fc1 = nn.Linear(32*32, 32)
        self.fc2 = nn.Linear(32, 8)
        self.fc3 = nn.Linear(8, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
#         x = self.ConvLayers(x)
        x = self.conv(x)
        x = x.view(-1, 32*32)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = self.sigmoid(x)
        return x


net = Net()
```
